```python
class TreeNode():
    def __init__(self,x):
        self.val=x
        self.left=None
        self.right=None
    def add(self,x):
            if x>self.val:
                if self.right!=None:
                    self.right.add(x)
                else:
                    self.right = TreeNode(x)
            else:
                if self.left!=None:
                    self.left.add(x)
                else:
                    self.left = TreeNode(x)      
```

```python
class printTree:
    def levelOrder(root):
            result=[]
            queue=[]
            if root:
                queue.append(root)
            while len(queue)!=0:
                layer=[]
                for _ in range(len(queue)):
                    x=queue.pop(0)
                    layer.append(x.val)
                    if x.left:
                        queue.append(x.left)
                    if x.right:
                        queue.append(x.right)
                result.append(layer)     
            return result
        
        
    def inOrder(root):
        res = []
        if root:
            res = inOrder(root.left) 
            res.append(root.val)
            res = res + inOrder(root.right)
        return res
        
```

```python
def convert(Str):
    listS=Str.split(" ")
    for i in range(len(listS)):
        listS[i]=int(listS[i])
    root=listS[0]
    t=TreeNode(root)
    
    for x in listS[1:]:
        t.add(x)
    yorder=printTree.levelOrder(t)
    xorder=printTree.inOrder(t)
    matrix = np.zeros((len(yorder),len(xorder)),dtype= object)
    
    for x in range(len(xorder)):
        element=xorder[x]
        for y in range(len(yorder)):
            try:
                yorder[y].index(element)
                matrix[y][x]=element
            except ValueError:
                pass
    
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j]==0.:
                matrix[i][j]='   '
            elif len(str(matrix[i][j]))==1:
                matrix[i][j]=str(matrix[i][j])+'  '
            elif len(str(matrix[i][j]))==2:
                matrix[i][j]=str(matrix[i][j])+' '
            else:
                matrix[i][j]=str(matrix[i][j])
            
    return matrix
```

```python
import numpy as np
fout=open('BTree.txt','w')
fin=open('input.txt','r',encoding="utf-8")

line=fin.readline()

while line:
    line=" ".join(line.split())
    output=convert(line)
    for row in output:
        for element in row:
            fout.write(element)
        fout.write('\n')
    fout.write('\n')
    
    line=fin.readline()

fin.close()
fout.close()
```
