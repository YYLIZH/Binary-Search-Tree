```python
def convert(Str):    
    class TreeNode():
        def __init__(self,x,posx,posy):
            self.val=x
            self.left=None
            self.right=None
            self.posx=posx
            self.posy=posy

        def add(self,x, px,py):
                if x>self.val:
                    if self.right!=None:
                        self.right.add(x,px+1,py+1)
                    else:
                        self.right = TreeNode(x,px,py)
                else:
                    if self.left!=None:
                        self.left.add(x,px-1,py+1)
                    else:
                        self.left = TreeNode(x,px,py)         

    listS=Str.split(" ")
    for i in range(len(listS)):
        listS[i]=int(listS[i])
    root=listS[0]
    t=TreeNode(root,0,0)
    for x in listS[1:]:
        t.add(x,0,0)

    '''
    leet 655
    https://blog.csdn.net/fuxuemingzhu/article/details/79439026
    '''
    class printTree:
        def plot(root):
            def getDepth(root):
                if not root:
                    return 0
                return 1+max(getDepth(root.left),getDepth(root.right))

            matrix=[]
            d=getDepth(root)
            cols=2**d-1
            for i in range(d):
                row=[]
                for j in range(cols):
                    row.append("")
                matrix.append(row)



            def set(root,d,row,column):
                matrix[row][column]=str(root.val)
                if root.right:
                    set(root.right,d-1,row+1,column+2**(d-2))
                if root.left:
                    set(root.left,d-1,row+1,column-2**(d-2))

            set(root,d,0,int((cols+1)/2)-1)

            return matrix
    return printTree.plot(t)
```

```python
fin=open('input.txt','r',encoding="utf-8")
fout=open('output002.txt','w')

line=fin.readline()

while line:
    line=" ".join(line.split())
    matrix=convert(line)
    for row in matrix:
        for element in row:
            if element == '':
                fout.write(' ')
            fout.write(element)
        fout.write('\n')
    fout.write('\n')
    fout.write('\n')
    
    line=fin.readline()
    
fin.close()
fout.close()
```
